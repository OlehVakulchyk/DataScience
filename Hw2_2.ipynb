Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"

```python
python_data_analysis = python_data.groupby(['Должность']).agg({'Зарплата.в.месяц': ['min', 'max']})
python_data_analysis
```

Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"

```python
def fill_avg_salary(x):
    fill_avg_salary = x.mean().round()
    return fill_avg_salary

python_data_analysis['avg'] = python_data_analysis.apply(fill_avg_salary, axis=1)
python_data_analysis['avg_mean'] = python_data.groupby(['Должность']).agg({'Зарплата.в.месяц': ['mean']}).round()
python_data_analysis
```

```python
import seaborn as sns
sns.set_style("darkgrid")

# Підготовка даних
# залишаємо тільки ті "Должность", кількість яких більше 6
df = d_s_without_nan.groupby('Должность').agg({'Зарплата.в.месяц': ['mean', 'count']}).round()
data = d_s_without_nan[d_s_without_nan['Должность'].isin(df[df['Зарплата.в.месяц','count']>6].index)]
# змінюємо та сортуємо дані за "Общий.опыт.работы"
data = data.replace(to_replace={'Общий.опыт.работы':{'10 и более лет': '>10', 'меньше 3 месяцев': '0.2'}})
data = data.sort_values('Общий.опыт.работы')

sns.lineplot(x='Общий.опыт.работы', y='Зарплата.в.месяц', hue='Должность', data=data)
```
